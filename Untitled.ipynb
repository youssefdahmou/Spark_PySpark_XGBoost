## 1) - Initialize PySpark:     

```python
from pyspark.sql import SparkSession
# Create a Spark session
spark = SparkSession.builder.appName("XGBoostExample").getOrCreate()
file_path = "data/stock_market_data.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)
```

```python
df.show(4)
```

## 2) - Feature Engineering: Prepare features and label for XGBoost.

```python
from pyspark.sql.functions import unix_timestamp
from pyspark.sql.types import DoubleType
from pyspark.ml.feature import VectorAssembler

# Convert 'Date' to timestamp and then to a numerical representation
df = df.withColumn("Date", unix_timestamp(df["Date"], "yyyy-MM-dd").cast(DoubleType()))

label_name = "Close"
feature_names = ["Date", "Open", "High", "Low", "Volume"]

# Rename the target variable
df = df.withColumnRenamed("Close", label_name)

# Create a VectorAssembler
assembler = VectorAssembler(inputCols=feature_names, outputCol="features")
df = assembler.transform(df)

# Show the transformed DataFrame
df.select(label_name, "features").show(truncate=False)
```

## 3) - Train-Test Split:

```python
train_df, test_df = df.randomSplit([0.8, 0.2], seed=42)
```

## 4) - Train XGBoost Model:

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from xgboost.spark import SparkXGBRegressor
```

```python
# Specify XGBoost parameters
xgb_params = {
    "objective": "reg:squarederror",
    "eta": 0.1,
    "maxDepth": 7,
    "subsample": 0.8,
    "colsampleByTree": 0.8,
    "numRound": 3000,
}

# Create XGBoostRegressor with SparkXGBRegressor
regressor = SparkXGBRegressor(
    features_col="features",
    label_col=label_name,
    numWorkers=2,
    **xgb_params
)
```

```python
# Train the model
model = regressor.fit(train_df)
```

## 5) - Make Predictions:

```python
# Make predictions on the test set
predictions = model.transform(test_df)
predictions.select(label_name, "prediction", "features").show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert the predictions DataFrame to Pandas for visualization
predictions_pd = predictions.select(label_name, "prediction").toPandas()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(predictions_pd[label_name], predictions_pd["prediction"], alpha=0.5)
plt.title("Actual vs Predicted")
plt.xlabel("Actual " + label_name)
plt.ylabel("Predicted " + label_name)
plt.show()
```

```python
def Data_Extractor(data):
    meta_data = data['Meta Data']
    symbol = meta_data['2. Symbol']
    last_refreshed = meta_data['3. Last Refreshed']
    
    # Extract time series data
    time_series_data = data['Time Series (5min)']
    
    # Create a DataFrame
    df_data = []
    
    for timestamp, values in time_series_data.items():
        row = {
            'Timestamp': timestamp,
            'Open': float(values['1. open']),
            'High': float(values['2. high']),
            'Low': float(values['3. low']),
            'Close': float(values['4. close']),
            'Volume': int(values['5. volume'])
        }
        df_data.append(row)
    return df_data 
```
